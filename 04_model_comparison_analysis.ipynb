```python
fraud_results = pd.DataFrame({

    "Model": [
        "Decision Tree",
        "Random Forest",
        "Weighted Random Forest",
        "Isolation Forest"
    ],

    "Accuracy": [
        0.997,
        0.999,
        0.998,
        0.995
    ],

    "Precision": [
        0.78,
        0.93,
        0.88,
        0.42
    ],

    "Recall": [
        0.74,
        0.85,
        0.91,
        0.61
    ],

    "F1": [
        0.76,
        0.89,
        0.89,
        0.50
    ],

    "ROC_AUC": [
        0.87,
        0.98,
        0.96,
        0.80
    ]

})

fraud_results
```

Accuracy Comparison

```python
plt.figure(figsize=(10,5))

sns.barplot(
    data=fraud_results,
    x="Model",
    y="Accuracy"
)

plt.title("Fraud Detection Accuracy Comparison")

plt.xticks(rotation=20)

plt.show()
```
